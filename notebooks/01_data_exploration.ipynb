```python
import matplotlib.pyplot as plt

with open("../data/cleaned/kannada_clean.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

lengths = [len(l.strip()) for l in lines[:100000]]
print(f"Total sentences: {len(lines):,}")
print(f"Avg length: {sum(lengths)/len(lengths):.0f} chars")
```

```python
plt.figure(figsize=(10, 4))
plt.hist(lengths, bins=50, edgecolor='black')
plt.xlabel("Sentence length (characters)")
plt.ylabel("Count")
plt.title("Distribution of Kannada Sentence Lengths")
plt.axvline(x=sum(lengths)/len(lengths), color='red', linestyle='--', label='Mean')
plt.legend()
plt.show()
```

```python
from collections import Counter
import re

kannada_chars = re.compile(r'[\u0C80-\u0CFF]')
all_chars = []
for line in lines[:50000]:
    all_chars.extend(kannada_chars.findall(line))

char_counts = Counter(all_chars)
top_20 = char_counts.most_common(20)
chars, counts = zip(*top_20)

plt.figure(figsize=(12, 4))
plt.bar(chars, counts)
plt.xlabel("Kannada character")
plt.ylabel("Frequency")
plt.title("Top 20 Most Common Kannada Characters")
plt.show()
```
